```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def plot_history(path,learning_rate,learning_rate2,k):
    last_values = {i:[] for i in ['accuracy','val_accuracy','test_accuracy','loss','val_loss','test_loss']}
    for metric in ['accuracy','loss']:
        for i in range(k):
            values = np.load(f"{path}history_{i}_{learning_rate}_{learning_rate2}_{metric}.npy")
            val_values = np.load(f"{path}history_{i}_{learning_rate}_{learning_rate2}_val_{metric}.npy")
            last_values[f"{metric}"].append(values[-1])
            last_values[f"val_{metric}"].append(val_values[-1])
            last_values[f"test_{metric}"].append(np.load(f"{path}value_{i}_{learning_rate}_{learning_rate2}_test_{metric}.npy"))

            plt.plot(values,'tab:blue')
            plt.plot(val_values,'tab:orange')

        plt.title(f"model {metric}")
        plt.ylabel(metric)
        plt.xlabel('epoch')
        if(metric=='accuracy'):
            plt.ylim([0.6,1])
        else:
            plt.ylim([0,1])
        plt.legend(['train', 'val'], loc='upper left')
        plt.show()
    
    matrix_test = np.mean([np.load(f"{path}value_{i}_{learning_rate}_{learning_rate2}_confusion_matrix.npy") for i in range(k)], axis=0)
    labels_x=['normal','excess','insufficient']
    labels_y = labels_x

    fig, ax = plt.subplots(figsize=(4, 3))
    ax.set_xlabel('Prédiction')
    ax.set_ylabel('Vraie étiquette')
    ax.set_title('Matrice de confusion test')
    print('Average test confusion matrix')
    sns.heatmap(matrix_test, annot=True, cmap="Greens", ax=ax, fmt='.0f', xticklabels=labels_x, yticklabels=labels_y)
    plt.show()
    
    for metric in ['accuracy','val_accuracy','test_accuracy','loss','val_loss','test_loss']:
        print(f"Moyenne {metric} : {np.mean(last_values[metric])}")
```

```python
path = 'raw_Adam_k=4_2e-05_5e-06/'
learning_rate= '2e-05'
learning_rate2='5e-06'
k=4

plot_history(path,learning_rate,learning_rate2,k)
```
