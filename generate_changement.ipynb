```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
def get_changeLuminosity_array(datasetLength):
    mean = datasetLength/2
    norm = scipy.stats.norm(mean, datasetLength*0.02)
    changeLuminosity = np.array([1-(norm.pdf(i)*datasetLength*0.02) for i in range(datasetLength)])
    # print(changeLuminosity)
    # changeLuminosity =  np.concatenate((np.ones(int(0.415*datasetLength)), changeLuminosity, np.ones(int(0.415*datasetLength))), axis=None)
    # plt.plot(changeLuminosity)
    # plt.show()
    return changeLuminosity
```

```python
## LOAD DATA
# Define the path to dataset folder
dataset_folder = 'subset/images/2023-04-25/3'
# dataset_folder = 'subset/test_change/'
os.makedirs(f'./generate_changement/2023-04-25/3')
partition = {}
# Iterate through each file in the dataset folder
train_partition = []
changeLuminosity = get_changeLuminosity_array(len(os.listdir(dataset_folder)))
for index,filename in enumerate(os.listdir(dataset_folder)):
    print(f'{dataset_folder}/{filename}')
    img = cv2.imread(f'{dataset_folder}/{filename}')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img[:,:,2] =img[:,:,2]*changeLuminosity[index]
    bright_img = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)
    cv2.imwrite(f'generate_changement/2023-04-25/3/{filename}', bright_img)
```
