```python
import numpy as np
import matplotlib.pyplot as plt
metrics = [{} for k in range(5)]
for k in range(5):
    first_history = np.load(f'./history/history_{k}.npy',allow_pickle='TRUE').item()
    fine_tuning_history = np.load(f'./history/history_{k}_fine_tuning.npy',allow_pickle='TRUE').item()
    for metric in ['accuracy','val_accuracy','loss','val_loss']:
        metrics[k][metric]=np.concatenate((first_history[metric],fine_tuning_history[metric]),axis=0)
```

```python
for k in range(5):
    plt.plot(metrics[k]['accuracy'],'tab:blue')
    plt.plot(metrics[k]['val_accuracy'],'tab:orange')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
for k in range(5):
    plt.plot(metrics[k]['loss'],'tab:blue')
    plt.plot(metrics[k]['val_loss'],'tab:orange')
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
means_final_metrics = {}
for metric in ['accuracy','val_accuracy','loss','val_loss']:
    means_final_metrics[metric] = np.mean([metrics[k][metric][len(metrics[k][metric])-1] for k in range(4)])
```
